```python
#Encryption function for a message in Z/pZ

def encrypt(p, pk, m):
    y = pk
    GF_q = FiniteField(q)
    r = GF_q.random_element()
    c1 = power_mod(g, r, p)
    d = power_mod(y, r, p)
    class_d = compute_class(d)
    c2 = (m + class_d) % p 
    return (c1, c2)

print(encrypt(p, pk, 4))
```
